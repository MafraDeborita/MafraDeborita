```python
df = pd.read_excel('segundo_semestre.xlsx')

# vizualizar primeiras linhas
df.head()
```

```python
# @title Indicador

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Indicador').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

### Filtrar apenas os indicadores de coleta de esgoto

```python
df_esgoto = df[df['Indicador'].str.contains('esgoto')]
```

### Ajustar os dados para o gráfico

```python
labels = df_esgoto['Indicador'].values
valores = df_esgoto['Valor'].values * 100  # transformar de proporção para percentual
```

```python
# Gráfico de pizza
plt.figure(figsize=(6,6))
plt.pie(valores,
        labels=labels,
        autopct='%1.1f%%',
        startangle=140,
        colors=['#4CAF50', '#F44336'])  # Verde e vermelho

plt.title('População com e sem coleta de esgoto - São Luís (2022)')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(labels, valores, color=['#4CAF50', '#F44336'])
plt.ylabel('% da população')
plt.title('Coleta de esgoto - São Luís (2022)')
plt.ylim(0, 100)
plt.show()

```

```python
df_agua = df[df['Indicador'].str.contains('água')]
```

### Filtrar apenas os indicadores de coleta de água

```python
df_agua = df[df['Indicador'].str.contains('água')]
```

### Ajustar os dados para o gráfico

```python
labels = df_agua['Indicador'].values
valores = df_agua['Valor'].values * 100  # transformar de proporção para percentual
```

```python
# Gráfico de pizza
plt.figure(figsize=(6,6))
plt.pie(valores,
        labels=labels,
        autopct='%1.1f%%',
        startangle=140,
        colors=['#4CAF50', '#F44336'])  # Verde e vermelho

plt.title('População com e sem coleta de água - São Luís (2022)')
plt.axis('equal')
plt.show()
```

### Filtrar indicadores relacionados à água

```python
df_agua = df[df['Indicador'].str.contains('água')].copy()
```

### Converter proporções (ex: 0.928) em porcentagens (ex: 92.8)

```python
df_agua['Valor (%)'] = df_agua['Valor'] * 100
```

```python
# Gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(df_agua['Indicador'], df_agua['Valor (%)'], color=['#4CAF50', '#F44336'])
plt.title('População com e sem acesso à água - São Luís (2022)')
plt.ylabel('% da população')
plt.ylim(0, 100)
#plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
